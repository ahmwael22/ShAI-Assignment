# Data_Cleaning

```python
df.info()
```

```python
df.drop_duplicates(inplace=True)
```

```python
df.drop(columns=["Notes","Status"],inplace=True)
```

```python
df.info()
```

```python
df.drop(columns="TotalPay",inplace=True)
```

```python
df['OvertimePay'].fillna(df['OvertimePay'].mean(), inplace=True)
```

```python
df['OtherPay'].fillna(df['OtherPay'].mean(), inplace=True)
```

```python
df['Benefits'].fillna(df['Benefits'].mean(), inplace=True)
```

```python

df.set_index("Id",inplace=True)

```

```python
df.drop(df[df["JobTitle"] == "Not provided"].index, inplace=True)
```

```python
 df[df["JobTitle"] == "Not provided"]["JobTitle"].value_counts()
```

# Basic Data Visualization 

```python
import matplotlib.pyplot as plt


plt.figure(figsize=(10, 6))
plt.hist(df['TotalPayBenefits'], bins=20, color='skyblue', edgecolor='black')
plt.title('Salary Distribution')
plt.xlabel('Salary')
plt.ylabel('Frequency')
plt.grid(True)
plt.show()
```
